```python
import numpy as np
from IPython.display import display, Markdown
```

```python
def f(X):
    return (X[2] + 2) * X[1] * X[0]**2

def g1(X):
    return 1 - ( X[1]**3 * X[2] )/( 71785 * X[0]**4 )

def g2(X):
    return ( 4 * X[1]**2 - X[0] * X[1] ) / ( 12566 * (X[1] * X[0]**3 - X[0]**4) ) + 1 / (5108 * X[0]**2) - 1

def g3(X):
    return 1 - ( 140.45 * X[0] ) / ( X[1]**2 * X[2] )

def g4(X):
    return ( X[0] + X[1] ) / 1.5 - 1
```

```python
def GeneratePop(PopSize):
    X1 = np.random.uniform(0.05, 2, [PopSize, 1])
    X2 = np.random.uniform(0.25, 1.30, [PopSize, 1])
    X3 = np.random.uniform(2.00, 15.0, [PopSize, 1])
    X = np.concatenate((X1, X2, X3), axis = 1)
    return X
```

```python
def Evaluate(X):
    if X[0] < 0.05 or X[0] > 2.00:
        #print("1")
        return False;
    if X[1] < 0.25 or X[1] > 1.30:
        #print("2")
        return False;
    if X[2] < 2.00 or X[2] > 15.0:
        #print("3")
        return False;
    if g1(X) > 0:
        #print("4")
        return False;
    if g2(X) > 0:
        #print("5")
        return False;
    if g3(X) > 0:
        #print("6")
        return False;
    if g4(X) > 0:
        #print("7")
        return False;
    return True;
```

```python
PopSize = 100000
InitPopulation = GeneratePop(PopSize)
```

```python
(Xa, fa) = (Xb, fb) = (Xd, fd) = (InitPopulation[0], f(InitPopulation[0]))
print("------------------   Population  ------------------")
Population = []
for X in InitPopulation:
    if Evaluate(X):
        if f(X) < fa:
            (Xd, fd) = (Xb, fb)
            (Xb, fb) = (Xa, fa)
            (Xa, fa) = (X, f(X))
        elif f(X) < fb:
            (Xd, fd) = (Xb, fb)
            (Xb, fb) = (X, f(X))
        elif f(X) < fd:
            (Xd, fd) = (X, f(X))
        print(X, f(X))
        Population.append(X)
print("------------------      TOP3     ------------------")
display(Markdown(f"$X_{{\\alpha}} = $ {Xa} | $ f(X_{{\\alpha}}) = $ {fa}"))
display(Markdown(f"$X_{{\\beta}} = $ {Xb} | $ f(X_{{\\beta}}) = $ {fb}"))
display(Markdown(f"$X_{{\\delta}} = $ {Xd} | $ f(X_{{\\delta}}) = $ {fd}"))
display(Markdown("---"))
print("Valid Population size =", len(Population))
```

```python
def CalculateNewX(X, a, i):
    r1 = np.random.uniform(0.0, 1.0)
    r2 = np.random.uniform(0.0, 1.0)
    A = 2 * a * r1 - a
    C = 2 * r2
    D = np.abs(C * X - Population[i])
    return X - A * D
    
```

```python
def Callibrate(Xnew):
    if Xnew[0] < 0.05:
        Xnew[0] = 0.05
    if Xnew[0] > 2.00:
        Xnew[0] = 2.00
    if Xnew[1] < 0.25:
        Xnew[1] = 0.25
    if Xnew[1] > 1.30:
        Xnew[1] = 1.30
    if Xnew[2] < 2.00:
        Xnew[2] = 2.00
    if Xnew[2] > 15.00:
        Xnew[2] = 15.00
```

```python
maxIter = 5
print("Population Size =", len(Population))
for Iter in range(1, maxIter + 1):
    display(Markdown("# Iteration N°%d"%Iter))
    for i in range(len(Population)):
        a = 2 * (1 - Iter / maxIter)
        X1 = CalculateNewX(Xa, a, i)
        X2 = CalculateNewX(Xb, a, i)
        X3 = CalculateNewX(Xd, a, i)
        Xnew = (X1 + X2 + X3) / 3
        Callibrate(Xnew)
        if f(Xnew) < f(Population[i]) and Evaluate(Xnew):
            print(Population[i], "with index", i, "has been updated with", Xnew)
            Population[i] = Xnew
            
    for X in Population:
        if f(X) < fa:
            (Xd, fd) = (Xb, fb)
            (Xb, fb) = (Xa, fa)
            (Xa, fa) = (X, f(X))
        elif f(X) < fb:
            (Xd, fd) = (Xb, fb)
            (Xb, fb) = (X, f(X))
        elif f(X) < fd:
            (Xd, fd) = (X, f(X))
    
    display(Markdown(f"$X_{{\\alpha}} = $ {Xa} | $ f(X_{{\\alpha}}) = $ {fa}"))
    display(Markdown(f"$X_{{\\beta}} = $ {Xb} | $ f(X_{{\\beta}}) = $ {fb}"))
    display(Markdown(f"$X_{{\\delta}} = $ {Xd} | $ f(X_{{\\delta}}) = $ {fd}"))
    display(Markdown("---"))

```
